# test version, try to clean column: student_name

```python
import os
import pandas as pd

student_data_to_load = os.path.join('Resources', 'students_complete.csv')
student_data_df = pd.read_csv(student_data_to_load)

student_data_df.head()
```

### use tolist() 

```python
# use _df[].tolist() seperate a column into a list
student_names = student_data_df["student_name"].tolist()
student_names[:10]
```

```python
# init three lists to append the wrong student names
students_to_fix=list()
prefixes=list()
suffixes=list()
for name in student_names:
    word=name.split()
    if len(word) >= 3:
        students_to_fix.append(name)
print(len(students_to_fix))
students_to_fix
```

```python
for name in students_to_fix:
            
    if len(name.split()[0]) <=4:
         prefixes.append(name.split()[0])
    if len(name.split()[-1]) <=3:  
         suffixes.append(name.split()[-1])
set(prefixes)

set(suffixes)
```

```python
#after set(), (like a only-value dictionary) use string method to replace() or strip()[not work well, removes letters]
#make a new list for ready to remove name
#return to DataFrame go through student_name column, replace(word, '')

prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

for word in prefixes_suffixes:
    student_data_df['student_name'] = student_data_df['student_name'].str.replace(word,'')
    
student_data_df.head(10)
```

### put cleaned student name in a new list, to double check

```python
#check len() of each name again
#now, if len greater than 3, means has family names, not prefix
students_fixed=list()
for name in student_names:
    word=name.split()
    if len(word) >= 3:
        students_fixed.append(name)
len(students_fixed)
#print(students_fixed)
```
